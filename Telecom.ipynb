```python
data['TotalCharges']=pd.to_numeric(data['TotalCharges'],errors='coerce')
```

```python
data.info()
```

```python
data1=data.drop(['customerID','SeniorCitizen','Dependents','PhoneService','PaymentMethod','Partner','PaperlessBilling','OnlineSecurity','DeviceProtection','StreamingTV','StreamingMovies'],axis=1)
```

```python
data1
```

```python
data2=data1.fillna(data1.median())
```

```python
data2
```

```python
data2['Churn']=data2['Churn'].map({'Yes':1,'No':0})
```

```python
data2
```

```python
data3=pd.get_dummies(data2)
```

```python
data3
```

```python
y=data3['Churn']
x=data3.drop('Churn',axis=1)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python

from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
